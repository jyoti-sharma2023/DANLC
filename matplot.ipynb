```python
import matplotlib.pyplot as plt
products=["iphone","samsung","oppo","vivo","oneplus"]
sales=[50000,34000,24000,30000,44000]
cs=["red","orange","green","blue","yellow"]
ex=[0.1,0,0.1,0,0.2]
textprops={
    "fontsize":12,
    'color':"purple",
    'weight': 'bold',
'font family': 'nonospace',
'fontstyle':'italic',
'rotation':30,
'backgroundcolor': 'blue',
'horizontalalignment':'center',
'verticalalignment': 'center',
'bbox':{'faceholder':'yellow','edgecolo':'red','pad':3}}
plt.pie(sales,labels=products, colors=cs,autopct="%1.1f%%",explode=ex,pctdistance=0.3,shadow=True, counterclock=False,wedgeprops={"edgecolor":"red",'linewidth': 2,"antialiased":False})
plt.legend()
plt.title("product wise sales data")
plt.show()
```

```python
import matplotlib.pyplot as plt
genres=["mystery","novel","fiction","fantasy","thriller"]
books_sold=[120,50,80,30,100]
ticks=["M","N","F","FY","T"]
C=["r","k","b","y","g"]
plt.bar(genres,books_sold ,align='center',color=C,edgecolor="r", linewidth=3,tick_label=ticks,width=0.5)

plt.xlabel("genres")
plt.ylabel("books_sold")

plt.show()
```

```python
import matplotlib.pyplot as plt
students=['riya','krisha','anjali','puja','sakshi']
maths_score=[76,78,88,98,80]
science_scores=[90,77,65,89,54]

f, plots=plt.subplots(1,2,figsize=(10,5))
plots[0].plot(students,maths_score,"red")
plots[0].plot(students,science_scores,"blue")
plots[0].set_xlabel("students")

plots[1].bar(students,maths_score,labels=students,color="g")
plots[1].bar(students,science_scores,labels=students,color="g")
plt.tight_layout()
plt.show()
```
